```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

plt.rcParams['font.family'] = ['MicroSoft YaHei']
plt.rcParams['axes.unicode_minus'] = False

df = pd.read_excel("CItest.xlsx")

# 
x = df['year']
y = df['data3']

n = len(y)

# 
# Skn0
Sk = np.zeros(n)
UFk = np.zeros(n)

# Sks
s = 0

for i in range(1, n):
    for j in range(0,i):
        if y.iloc[i] > y.iloc[j]:
            s += 1
    Sk[i] = s
    E = (i+1)*(i/4)
    Var = (i+1)*i*(2*(i+1)+5)/72
    UFk[i] = (Sk[i] - E)/np.sqrt(Var)
    
# 
# Sk2
# Sk2
y2 = np.zeros(n)
Sk2 = np.zeros(n)
UBk = np.zeros(n)

s = 0
y2 = y[::-1]

for i in range(1, n):
    for j in range(0,i):
        if y2.iloc[i] > y2.iloc[j]:
            s += 1
    Sk2[i] = s
    E = (i+1)*(i/4)
    Var = (i+1)*i*(2*(i+1)+5)/72
    UBk[i] = -(Sk2[i] - E)/np.sqrt(Var)

UBk2 = UBk[::-1]

# 
plt.figure(figsize=(7, 6), dpi=350)
plt.plot(range(30,),UFk, label='UF', color='black',marker='s')
plt.plot(range(30,), UBk2, label='UB',color='black', linestyle='--', marker='o')
plt.ylabel('Mann-Kendall ')
plt.xlabel(' Year')

# 
x_lim = plt.xlim()
# y=0
plt.plot(x_lim,[-1.96,-1.96],':',color='black',label='5%')
plt.plot(x_lim, [0,0],'--',color='black')
plt.plot(x_lim,[1.96,1.96],':',color='black')
plt.xticks(range(30), x.tolist(), rotation=90)
# plt.legend(loc='upper right', bbox_to_anchor=(0.9,0.95),ncol=3,fancybox=True)

# 
plt.legend(bbox_to_anchor=(0.3,0.9), facecolor='w',frameon=False)
# 
# plt.text(-0.9,2.25,'')
plt.savefig("MK.png")
plt.show()
```
